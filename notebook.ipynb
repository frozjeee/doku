```python
import pandas as pd
import matplotlib.pyplot as plt
import os


class PltDrawer:
    output_folder = "plots"
    def __init__(self, json_file):
        self.df = pd.read_json(json_file)
        
    def draw_plots(self):
        plots = []
        
        self.df['corner_diff'] = self.df['rb_corners'] - self.df['gt_corners']
        self.df['corner_diff'].hist(bins=30, color='skyblue', edgecolor='black')
        plt.title('Histogram of Corner Difference (Predicted - Ground Truth)')
        plt.xlabel('Difference in Corners')
        plt.ylabel('Frequency')
        plot_path = 'corner_diff.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()


        self.df['mean'].hist(bins=30, color='skyblue', edgecolor='black')
        plt.title('Histogram of Mean Deviations')
        plt.xlabel('Mean Deviation')
        plt.ylabel('Frequency')
        plot_path = 'mean_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['max'].hist(bins=30, color='salmon', edgecolor='black')
        plt.title('Histogram of Max Deviations')
        plt.xlabel('Max Deviation')
        plt.ylabel('Frequency')
        plot_path = 'max_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['min'].hist(bins=30, color='lightgreen', edgecolor='black')
        plt.title('Histogram of Min Deviations')
        plt.xlabel('Min Deviation')
        plt.ylabel('Frequency')
        plot_path = 'min_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()


        self.df['ceiling_mean'].hist(bins=30, color='skyblue', edgecolor='black')
        plt.title('Histogram of Ceiling Mean Deviations')
        plt.xlabel('Mean Deviation')
        plt.ylabel('Frequency')
        plot_path = 'ceiling_mean_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['ceiling_max'].hist(bins=30, color='salmon', edgecolor='black')
        plt.title('Histogram of Ceiling Max Deviations')
        plt.xlabel('Max Deviation')
        plt.ylabel('Frequency')
        plot_path = 'ceiling_max_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['ceiling_min'].hist(bins=30, color='lightgreen', edgecolor='black')
        plt.title('Histogram of Ceiling Min Deviations')
        plt.xlabel('Min Deviation')
        plt.ylabel('Frequency')
        plot_path = 'ceiling_min_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()


        self.df['floor_mean'].hist(bins=30, color='skyblue', edgecolor='black')
        plt.title('Histogram of Floor Mean Deviations')
        plt.xlabel('Mean Deviation')
        plt.ylabel('Frequency')
        plot_path = 'floor_mean_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['floor_max'].hist(bins=30, color='salmon', edgecolor='black')
        plt.title('Histogram of Floor Max Deviations')
        plt.xlabel('Max Deviation')
        plt.ylabel('Frequency')
        plot_path = 'floor_max_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        self.df['floor_min'].hist(bins=30, color='lightgreen', edgecolor='black')
        plt.title('Histogram of Floor Min Deviations')
        plt.xlabel('Min Deviation')
        plt.ylabel('Frequency')
        plot_path = 'floor_min_deviation_per_room.png'
        plt.savefig(os.path.join(self.output_folder, plot_path))
        plots.append(plot_path)
        plt.close()

        return plots
```

```python
drawer = PltDrawer("deviation.json")
drawer.draw_plots()
```
